```python
#6. Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.
```

```python
hare_c=np.diff(hare)
lynx_c = np.diff(lynx)
carrot_c=np.diff(carrot)

plt.figure(figsize=(12,5))
plt.plot(x[1:],hare_c, label='Hare change')
plt.plot(x[1:],lynx_c, label='Lynx change')
plt.xlabel('Year')
plt.ylabel('Population change')
plt.title('Yearly changes in hare and lynx populations')
plt.xticks(x[1:])
plt.grid(True)
plt.legend()
plt.show()

hd  = np.diff(hare)
ld = np.diff(lynx)

mean_h = np.mean(hd)
mean_l = np.mean(ld)

cov = np.sum((hd - mean_h) * (ld - mean_l)) / (len(hd) - 1)

std_h = np.sqrt(np.sum((hd - mean_h)**2) / (len(hd) - 1))
std_l = np.sqrt(np.sum((ld - mean_l)**2) / (len(ld) - 1))

cor = cov / (std_h * std_l)

print("\nCorrelation =", cor)
```

```python
# :

# 1. ,   k-means.
#    

#  NxM,  N --  , M --   
# k --  
# max_iter --   
# threshold --   (         ,   threshold,  )
```

```python
def k_means(arr, k, max_iter, threshold):
    N, M = arr.shape
    centers = arr[:k].copy()
    index = np.arange(N)    
    sse = 0
    for i in range(max_iter):
        #
        for j in range(N):
            values = np.array([np.linalg.norm(arr[j] - centers[p]) for p in range(k)])
            index[j] = np.argmin(values)
        
        #  
        for l in range(k):
            mask  = (index == l)
            centers[l] = arr[mask].mean(axis=0)

        # :
        sse1 = 0
        for s in range(N):
            sse1 += np.linalg.norm(arr[s] - centers[index[s]])**2
        if(i!=0):
            if(abs(sse -sse1)<threshold):
                sse = sse1
                break
            else:
                sse = sse1
        else:
            sse =sse1

    return index, sse, centers
    

        
```

```python
from sklearn.datasets import fetch_openml
images, labels = fetch_openml('mnist_784', return_X_y=True, as_frame=False, parser = "auto")
```

```python
from sklearn.metrics import confusion_matrix

labels = labels.astype(int)  
images = images / 255.0

k = 10
max_iter = 10  
threshold = 0

index, sse, centers = k_means(images, k, max_iter, threshold)
print(" SSE:", sse)


fig, axes = plt.subplots(1, 10, figsize=(15, 3))
for i, ax in enumerate(axes):
    ax.imshow(centers[i].reshape(28, 28), cmap='Greys_r')
    ax.set_title(f'Cluster {i}')
    ax.axis('off')
plt.suptitle("  k-means")
plt.show()

cluster_to_digit = {}
print("\n :")
for c in range(k):
    cluster_mask = (index == c)
    cluster_labels = labels[cluster_mask]
    if cluster_labels.size == 0:
        print(f" {c}: ")
        continue

    values, counts = np.unique(cluster_labels, return_counts=True)
    majority_digit = values[np.argmax(counts)]
    cluster_to_digit[c] = majority_digit

    print(f" {c}:    {majority_digit} "
          f"({np.max(counts)}  {cluster_labels.size})")


print("\n     :")
for c, digit in cluster_to_digit.items():
    mask = (index == c)
    cluster_labels = labels[mask]
    correct = np.sum(cluster_labels == digit)
    total = cluster_labels.size
    acc = correct / total if total > 0 else 0
    print(f" {c}   {digit}: {correct}/{total} "
          f"({acc:.2%} )")


predicted_digits = np.zeros_like(labels)
for c, digit in cluster_to_digit.items():
    predicted_digits[index == c] = digit

cm = confusion_matrix(labels, predicted_digits, labels=np.arange(10))

print("\n  (   ,   ):\n")

for i in range(10):
    row = " ".join([f"{cm[i, j]:5d}" for j in range(10)])
    print(f"{i:2d} | {row}")
```
